```python
proteomics['ID_Visit']=None
```

```python
for i in range(len(proteomics)):
    if proteomics['Month'][i]==0.0:
        proteomics['ID_Visit'][i]=proteomics['Unnamed: 0'][i].split('_')[0]+'_m00'
    elif proteomics['Month'][i]==12.0:
        proteomics['ID_Visit'][i]=proteomics['Unnamed: 0'][i].split('_')[0]+'_m12'
```
